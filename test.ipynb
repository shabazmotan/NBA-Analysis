```python
import pandas as pd 
```

```python
Reg22_df=pd.read_csv('RegSeason22.csv', encoding= 'latin1', delimiter=';')
Reg22_df
```

```python
Playoff22_df=pd.read_csv('Playoff22.csv', encoding= 'latin1', delimiter=';')
Playoff22_df
```

```python
Reg22_df
```

```python
# TS%= PTS/(2*TSA)
# TSA=FGA+(.44*FTA)
Reg22_df['TSA']=(Reg22_df['FGA']+(.44*Reg22_df['FTA']))
Reg22_df
```

```python
Reg22_df['TS%']=(Reg22_df['PTS']/(2*Reg22_df['TSA']))
```

```python
Playoff22_df['TSA']=(Playoff22_df['FGA']+(.44*Playoff22_df['FTA']))
Playoff22_df
```

```python
Playoff22_df['TS%']=(Playoff22_df['PTS']/(2*Playoff22_df['TSA']))
```
