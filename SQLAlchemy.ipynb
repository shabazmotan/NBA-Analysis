```python
# Connect to SQLite
from sqlalchemy import create_engine, Column, Integer, String, Float
from sqlalchemy.orm import Session
from sqlalchemy.ext.declarative import declarative_base
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.sql import func
from sqlalchemy import desc
import sqlalchemy
```

```python
engine = create_engine("postgresql://postgres:postgres@localhost:5432/Project3")
conn=engine.connect()
session=Session(bind=engine)
session
```

```python
engine.execute('SELECT * from RegSea_22;').fetchall()
```

```python
Base=automap_base()
```

```python
Base.prepare(autoload_with=engine)
```

```python
Base.classes.keys()
```

```python
RegSeason=Base.classes.regsea_22
```

```python
players=session.query(RegSeason).all()
```

```python
for player in players:
    print(f'{player.player}, {player.pos}, {player.team}, {player.gamesplayed}, {player.minutesplayed}')
```

```python
# League Avg TS% for the season was 52.7%.
# Above Average Scorers
Efficient_Scorers=session.query(RegSeason).\
    filter(RegSeason.ts_per>0.527).\
    filter(RegSeason.gamesplayed>41).\
    filter(RegSeason.team!='TOT').\
    filter(RegSeason.fga>5)
for scorer in Efficient_Scorers:
    print(f'{scorer.player}, {scorer.ts_per}')
```

```python
# Select avg(regsea_22.threeper) as average from regsea_22
# Where threepa>3
# Efficient Three-PtShooters
Efficient_Shooters=session.query(RegSeason).\
    filter(RegSeason.ts_per>0.527).\
    filter(RegSeason.gamesplayed>41).\
    filter(RegSeason.team!='TOT').\
    filter(RegSeason.fga>5).\
    filter(RegSeason.threepa>3.5).\
    filter(RegSeason.threeper>.354)
for shooter in Efficient_Shooters:
    print(f'{shooter.player}, {shooter.ts_per}')
```

```python
# Efficient Scorers at the Line
Efficient_FT=session.query(RegSeason).\
    filter(RegSeason.ts_per>0.527).\
    filter(RegSeason.gamesplayed>41).\
    filter(RegSeason.team!='TOT').\
    filter(RegSeason.fta>4).\
    filter(RegSeason.ft_per>0.795)
for shooter in Efficient_FT:
    print(f'{shooter.player}, {shooter.ts_per}, {shooter.ft_per}')
```

```python
# Good Defenders by metrics
Defensive_Players=session.query(RegSeason).\
    filter(RegSeason.gamesplayed>41).\
    filter(RegSeason.team!='TOT').\
    filter(RegSeason.stl>.999).\
    filter(RegSeason.blk>.499)
for defender in Defensive_Players:
    print(f'{defender.player},{defender.minutesplayed}, {defender.stl}, {defender.blk}')
```

```python

Shot_Blockers=session.query(RegSeason).order_by(desc(RegSeason.blk)).\
    filter(RegSeason.gamesplayed>41).\
    filter(RegSeason.team!='TOT')

for blocks in Shot_Blockers:
    print(f'{blocks.player}, {blocks.blk}')
```
